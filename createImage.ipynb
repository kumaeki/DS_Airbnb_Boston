```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df_listing_b_origin = pd.read_csv("./data/Boston/listings.csv")
```

```python
# Creating word_cloud with text as argument in .generate() method
def create_word_cloud(df, df_name, col_name):
    txt = " ".join(nbhd for nbhd in df[col_name].fillna(""))
    image_fn = "{}_{}.png".format(df_name, col_name)

    wc = WordCloud(
        collocations=False, background_color="white", width=2048, height=1080
    ).generate(txt)
    wc.to_file(image_fn)

    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.show()


create_word_cloud(df_listing_b_origin, "df_listing_b_origin", "neighborhood_overview")
```
